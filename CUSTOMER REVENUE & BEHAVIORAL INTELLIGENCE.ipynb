```python
import pandas as pd

df_raw = pd.read_csv("customer_shopping_behavior.csv")
df = df_raw.copy()

df.to_csv("customer_behavior_copy.csv", index=False)
```

```python
df.shape
df.info()
df.isnull().sum()
```

```python
# 1. Customer identity
id_cols = [
    "Customer ID"
]

# 2. Target (what business cares about)
target_cols = [
    "Purchase Amount (USD)"
]

# 3. Customer behavior (numeric, continuous)
num_cols = [
    "Age",
    "Review Rating",
    "Previous Purchases"
]

# 4. Promotion & subscription flags (Yes/No business signals)
flag_cols = [
    "Discount Applied",
    "Promo Code Used",
    "Subscription Status"
]

# 5. Customer & product attributes (dimensions for Power BI slicing)
attribute_cols = [
    "Gender",
    "Item Purchased",
    "Category",
    "Location",
    "Size",
    "Color",
    "Season",
    "Shipping Type",
    "Payment Method",
    "Frequency of Purchases"
]
```

```python


for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
df[flag_cols] = df[flag_cols].fillna("No")
df.isnull().sum()
```

```python
df["Age"] = df["Age"].astype(int)
df["Previous Purchases"] = df["Previous Purchases"].astype(int)
df["Review Rating"] = df["Review Rating"].round(1)
```

```python
df.head(10)
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
bins = [0, 24, 34, 49, 100]
labels = ['Young', 'Young Adult', 'Adult', 'Senior']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
# Repeat Customer (1 = repeat, 0 = first time)
df["is_repeat_customer"] = (df["previous_purchases"] > 0).astype(int)
```

```python
def loyalty_segment(p):
    if p == 0:
        return "New"
    elif p <= 3:
        return "Occasional"
    elif p <= 10:
        return "Regular"
    else:
        return "Loyal"

df["loyalty_segment"] = df["previous_purchases"].apply(loyalty_segment)
```

```python
df["spend_band"] = pd.cut(
    df["purchase_amount"],
    bins=[0, 50, 100, 200, 100000],
    labels=["Low", "Medium", "High", "Very High"]
)
```

```python

df["estimated_clv"] = (
    df["purchase_amount"] *
    (df["previous_purchases"] + 1)
).round(2)

```

```python
df["is_vip"] = ((df["estimated_clv"] > 1000) & (df["previous_purchases"] >= 5)).astype(int)
```

```python
df["discount_used_flag"] = (df["discount_applied"] == "Yes").astype(int)
```

```python
# After all cleaning / feature engineering is done
# (after age_group, purchase_frequency_days etc are created)

cleaned_filename = "customer_behavior_cleaned.csv"

# Save exactly what is currently in df to disk
df.to_csv(cleaned_filename, index=False)

print("Saved file:", cleaned_filename)
print("Columns saved:")
print(df.columns.tolist())
print("Rows:", len(df))
```
